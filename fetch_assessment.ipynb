```python
import pandas as pd
from sqlalchemy import create_engine
import json
```

```python
receipts = pd.read_json('receipts.json', lines=True)
brands = pd.read_json('brands.json', lines=True)
users = pd.read_json('users.json', lines=True)
```

```python
receipts.head()
```

```python
receipts_df = []
receipt_item_list_df = []
# new receipts_df dataframe
for receipt in receipts['_id']:
    receipt_create_date = receipts['createDate']
    rewards_receipt_status = receipts['rewardsReceiptStatus']
    spend = receipts['totalSpent']
    user_id = receipts['userId']
    receipt_id = receipts['_id'][0]
    receipts_df.append({'id': receipt_id, 'date': receipt_create_date, 'status': rewards_receipt_status,
                       'spend': spend, 'user_id': user_id})
# Receipt Item List dataframe -- joined to receipts_df dataframe on key RIL_receipt_id = receipt_id
    for i in receipts['purchasedItemCount']:
        RIL_receipt_id = receipts['_id'][0]
        barcode = receipts['rewardsReceiptItemList'][i]['userFlaggedBarcode']
        item_description = receipts['rewardsReceiptItemList'][i]["userFlaggedDescription"]
        partner_id = receipts['rewardsReceiptItemList'][i]["rewardsProductPartnerId"]
        receipt_item_list_df.append({'id': RIL_receipt_id, 'barcode': barcode, 'item': item_description, 
                                     'partner': partner_id})
```

```python
brands_list = []
for brand in brands['_id']:
    category = brands['category']
    top_brand = brands['topBrand']
    name = brands['name']
    brand_id = brands['cpg'][1]
    brands_list.append({'id': brand_id, 'top brand': top_brand, 'category': category, 'name': name})
```

```python
users_list = []
for user in users['_id']:
    user_create_date = users['createdDate'][0]
    active = users['active']
    last_login = users['lastLogin'][0]
    user_id = users['_id'][0]
    state = users['state']
    users_list.append({'id': user_id, 'date': user_create_date, 'active': active, 'state': state,
                       'last login': last_login})
```

```python
brands_df = pd.DataFrame(brands_list)
brands_df.head()
```
